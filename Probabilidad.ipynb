## **IF-ELSE STATEMENT**

### ***If Statement:***
-The if statement is used to establish a condition that must be fulfilled to run a part of the code. We include a colon (**:**) at the end of the line to tell Python we are creating a block of code, and then indenting four spaces tells Python what lines of code are part of the block.

```python
Mustang = 700
Camaro = 500
if Mustang > Camaro : #Condition
    print("Mustang is most better than Camaro") #Code that will be run if the condition is fulfilled.
```

```python
Mustang = 700
Challenger = 600
Camaro = 500
if Mustang >= Challenger and Camaro < Challenger:  #Mustang must be greater than or equal to Challenger, and Camaro must be less than Challenger.
    print("Mustang is the best Muscle Car")
```

-If only one of the conditions must be fulfilled, write "**or**" between the conditions.

```python
Mustang = 700
Challenger = 600
Camaro = 500
if Mustang >= Challenger or Camaro < Challenger:  #Mustang must be greater than or equal to Challenger, or Camaro must be less than Challenger.
    print("Challenger at least is better than one of them")
```

```python
age = int(input("What model is your car? "))
if age <= 1999:
    print("Wow you have a classic car!!!")
if age > 2000:
    print("Cool Bro! jeje")
```

### ***Else Statement:***
-The ***else*** statement is used to run a part of the code in case the condition of the ***if*** statement is not fulfilled.

```python
Mustang = 700
Challenger = 600
if Mustang > Challenger :
  print("Mustang is better than Challenger")
else:
  print("The challenger is modified")
```

## **Loops Statement**

### ***For Loop:***
-A loop is a part of the code that will be run many times until a condition is fulfilled.

-A For loop use a variable that increase its value by 1 every time the code inside it is ran. When this variable reach the limit value, the loop stops and Python continues with the next part of the code.
The structure of a for loop is:

***for*** name_of_the_variable ***in range(***first_value,last_value***):***

```python
for i in range(0,6): #The loop will be run 6 times
    print("I Love Mustangs <3", end = '')
```

-Instead of use a range, you can use the name of a list in the For loop. In this case, the variable will take the values inside the list one by one.

-***REMEMBER*** end = ' ' is used to continue writing in the same line instead of jumping to the next one.

```python
Model = ["Cobra", "Shelby", "EcoBoots", "Boss", "Mach 1"]
for Car in Model:
    print(Model, end = '')
```

### ***While Loop:***
-A while loop will keep running until the condition in it is ***FALSE***.

```python
y = 0
x = []
while y < 1: #The condition for the loop is y less than 1
    print("Write 0 to stop")
    number = int(input("Give me a number: ")) #Ask for a number and save it in the variable number
    x.append(number) #Save the variable number in the list x
    
    if number == 0: #If the user introduces a 0, the value of y increases and the condition of the loop becomes FALSE
        y += 1
print(x)
```
